**Checking most occuring time frame on time basis**

```python
import pandas as pd

# Creating a function to format time
def format_time_range(hour):
    try:
        start_time = pd.to_datetime(f'{int(hour)}:00', format='%H:%M').strftime('%I:00 %p').lstrip('0')
        end_time = pd.to_datetime(f'{(int(hour) + 1) % 24}:00', format='%H:%M').strftime('%I:00 %p').lstrip('0')
        return f'{start_time} - {end_time}'
    except ValueError:
        return "Invalid Time"

# Extracting the hour from CallTime
df_cleaned['Hour'] = df_cleaned['CallTime'].dt.hour  # Ensure 'CallTime' is in datetime format

# Counting the number of incidents for each hour
hourly_counts = df_cleaned['Hour'].value_counts().sort_index()

# Formatting the hour into time ranges for better readability
hourly_counts.index = hourly_counts.index.map(format_time_range)

# Finding the most common time frame
most_common_hour = hourly_counts.idxmax()
highest_accident_count = hourly_counts.max()

# Output results
print(f"Time frame with the most accidents: {most_common_hour}")
print(f"Number of accidents in that time frame: {highest_accident_count}")
print("\nAccident count by time frame:")
print(hourly_counts)
```
